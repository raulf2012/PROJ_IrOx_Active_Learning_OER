```python
['001', '010', '100', '110', '111', '211']
```

```python
format_dict = {}
for i in smart_format_dict:
    format_dict[i[0]["facet"]] = i[1]["dash"]

data = []
for ind, i in enumerate(format_dict.items()):
    trace_i = go.Scatter(
        x = [0, 0.2], y = [ind + 1., ind + 1.],
        mode = 'lines',
        name = i[0],
        line = dict(
            dash=i[1],
            width=4.,
            color="black",
            ),
        )
    data.append(trace_i)
```

```python
trace_i = go.Scatter(
    x = [0, 0.2], y = [0., 0.],
    mode = 'lines',
    name = i[0],
    line = dict(
#         dash="solid",
        width=1.,
        color="black",
        ),
    )
data.append(trace_i)
```

```python
layout = {
    "width": 100 * 2.5,
    "height": 100 * 3.,
    "showlegend": False,
    }

fig = go.Figure(data=data, layout=layout)
```

```python
py.iplot(fig,
    filename=os.path.join(
        proj_dir_name,
        "surface_pourbaix",
        "line_type_legend",
        )
    )
```
